```python
import pandas as pd
import requests
```

## Dataset releated to influenza researches

```python
influenzaResearchDB_URL_QUERY = "datatype=genome&family=influenza&country=China&fromyear=2016&toyear=2016&subtype=H3N2&segment=4&output=json"

influenzaResearchDB_REQUEST_URL = "https://www.fludb.org/brc/api/sequence?"

influenzaData_AS_JSON = requests.get(influenzaResearchDB_REQUEST_URL, params=influenzaResearchDB_URL_QUERY).content
```

### The json result is too big to show, so we can convert it to a pandas dataframe and better visualize the data

```python
influenzaResearchesDataset = pd.read_json(influenzaData_AS_JSON)

influenzaResearchesDataset
```
